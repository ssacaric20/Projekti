```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Podijela na numericke i kategorijske znacajke
numericke = [
    'Floor_Insulation_U-Value', 'Door_Insulation_U-Value', 'Roof_Insulation_U-Value',
    'Window_Insulation_U-Value', 'Wall_Insulation_U-Value', 'HVAC_Efficiency',
    'Domestic_Hot_Water_Usage', 'Lighting_Density', 'Occupancy_Level',
    'Equipment_Density', 'Heating_Setpoint_Temperature', 'Heating_Setback_Temperature',
    'Air_Change_Rate', 'Window_to_Wall_Ratio', 'Total_Building_Area', 'Heating_Usage',
    'Water_Systems_Energy', 'Interior_Lighting_Energy', 'Interior_Equipment_Energy',
    'Total_Heating_Energy', 'Photovoltaic_Power', 'Total_Electricity_Energy',
    'Heating_Conversion_Factor', 'Electricity_Primary_Conversion_Factor',
    'Heating_Primary_Conversion_Factor', 'Energy_Use_Intensity'
]

kategorijske = [
    'Simple_Building_Energy_Rating', 'Weather_File', 'Building_Type',
    'Renewable_Energy_Usage', 'Building_Energy_Rating'
]
```

```python
# Skaliranje numerickih znacajki
scaler = StandardScaler()
scaled_data = scaler.fit_transform(urban_dataset[numericke])
scaled_df = pd.DataFrame(scaled_data, columns=numericke)

# Pretvorba kategorijskih vrijednosti u numericke
encoded_df = pd.get_dummies(urban_dataset[kategorijske], dtype=int)

# Kombiniranje oba dataframea
final_df = pd.concat([scaled_df, encoded_df], axis=1)
```

```python
#from sklearn.decomposition import PCA

# Uzorkovanje podataka (5% originalnog skupa podataka)
sampled_df = final_df.sample(frac=0.05, random_state=42)

#sampled_dataset = urban_dataset.loc[sampled_df.index] Ehh failed pokusaj
```

```python
# Primjena Kmeans algoritma

# Vizualizacija preko metode laktova za optimalan broj klastera
inercija = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(sampled_df)
    inercija.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inercija, 'bo-', markersize=8)
plt.xlabel('Broj klastera (K)')
plt.ylabel('Inercija')
plt.title('Metoda lakta a optimalan broj K')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Raspon brojeva klastera za analizu
k_values = [3,4,5]
silhouette_scores = []

# Izračun Silhouette Score za svaki broj klastera
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(sampled_df)
    score = silhouette_score(sampled_df, clusters)
    silhouette_scores.append(score)
    print(f"K={k}: Silhouette Score={score:.4f}")

# Identifikacija najboljeg K
best_index = silhouette_scores.index(max(silhouette_scores))
best_k = k_values[best_index]
best_score = silhouette_scores[best_index]

print(f"Optimalan broj klastera prema Silhouette Score je K = {best_k} sa scoreom = {best_score:.4f}")

# Za izračun vrijednosti potrebno je malo duže pričekati zbog velike količine podataka
```

```python
# Distribucija podataka po klasterima za najbolji broj klastera
kmeans = KMeans(n_clusters=best_k, random_state=42)
clusters = KMeans(n_clusters=best_k, random_state=42).fit_predict(sampled_df)
sampled_df['Klaster'] = clusters
cluster_counts = sampled_df['Klaster'].value_counts()

# Graf distribucije podataka po klasterima
plt.figure(figsize=(8, 5))
cluster_counts.sort_index().plot(kind='bar', color='pink')
plt.xlabel('Klasteri')
plt.ylabel('Broj podataka')
plt.title(f'Distribucija podataka po klasterima (K={best_k})')
plt.show()
```

```python
# PCA metoda za smanjenje dimenzionalnosti
pca = PCA(n_components=2, random_state=42)  # Reduciranje na dvije dimenzije
reduced_data = pca.fit_transform(sampled_df)

# Dodavanje pca1 i pca2 kao dvije nove znacajke u set
sampled_df['pca1'] = reduced_data[:, 0]
sampled_df['pca2'] = reduced_data[:, 1]

# Scatterplot za prikaz klastera i rasprsenosti podataka
plt.figure(figsize=(10, 7))
sns.scatterplot(data=sampled_df, x='pca1', y='pca2', hue=sampled_df['Klaster'], palette='viridis', s=50)
plt.title(f'Klasteri (K={best_k}) (reducirano PCA metodom)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
